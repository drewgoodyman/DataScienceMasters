```python
# 1a. Read in the aapl.csv, amzn.csv, fb.csv, goog.csv, and nflx.csv files.
import pandas as pd 
# Reads each csv file and saves data as a variable
aapl = pd.read_csv('aapl.csv')
amzn = pd.read_csv('amzn.csv')
fb = pd.read_csv('fb.csv')
goog = pd.read_csv('goog.csv')
nflx = pd.read_csv('nflx.csv')
```

```python
# 1b. Add a column to each dataframe, called ticker, indicating the ticker symbol it is for (Apple's is AAPL, for example);
# this is how you look up a stock. In this case, the filenames happen to be the ticker symbols.

# Adds a column called 'ticker' to each data frame with its own ticker attached i.e. AAPL for Apple, AMZN for Amazon, etc.
aapl['ticker'] = 'AAPL'
amzn['ticker'] = 'AMZN'
fb['ticker'] = 'FB'
goog['ticker'] = 'GOOG'
nflx['ticker'] = 'NFLX'

# Displays aapl columns to display that the column ticker was added
aapl.columns
```

```python
# 1c. Append them together into a single dataframe.
# Combines the fives data frames into a single dataframe and ignores the original indicies
faang = pd.concat([aapl, amzn, fb, goog, nflx], ignore_index = True)
```

```python
# 1da. Displays the first few rows of faang data frame.
faang.head()
```

```python
# 1db. Displays the last few rows of faang data frame.
faang.tail()
```

```python
# 1e. e) Save the result in a CSV file called faang.csv.
# Saves the faang dataframe into a csv without saving an index column
faang.to_csv('faang.csv', index=False)
```

```python
# The European Centre for Disease Prevention and Control (EDC) provides an open dataset on COVID-19 cases called daily
# number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot
# that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format:
```

```python
# Load the COVID-19 csv file
covid = pd.read_csv('covid19_cases.csv')
```

```python
# View current column names
covid.columns
```

```python
# Create a 'date' column from the dateRep
covid['date'] = pd.to_datetime(covid['dateRep'], dayfirst=True)
```

```python
# Set 'date' as the index then sort the data frame by the index
covid.set_index('date', inplace=True)
covid.sort_index(inplace=True)
```

```python
# View updated data frame to show 'date'
covid.head()
```

```python
# Replace all occurrences of United_States_of_America and United_Kingdom with USA and UK
covid = covid.replace({'United_States_of_America': 'USA', 'United_Kingdom': 'UK'})
```

```python
# Define a list of countries to keep data on
countries = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA']
```

```python
# Filter the covid data frame for the countries using the 'countriesAndTerritories' colum
covid_filtered = covid[covid['countriesAndTerritories'].isin(countries)]
```

```python
# Pivot the data so that:
# - The index is the dates
# - The columns are the country names 
# - The values are the 'cases' counts 
covid_pivot = covid_filtered.pivot_table(index=covid_filtered.index,
                                         columns='countriesAndTerritories',
                                         values='cases').fillna(0)
```

```python
# Print the covid pivot data table
covid_pivot
```
